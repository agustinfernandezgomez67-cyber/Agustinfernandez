```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Data/cobros_juan.csv", sep = ",")
```

```python
df.count()
```

```python
Sueldo_promedio = df['Cobro_USD'].mean()
print("\n--- Resultado del Cálculo ---")
print(f"El sueldo promedio mensual de Juan fue de: ${Sueldo_promedio:.2f}")
```

```python
def clasificar_sueldo(cobro):
    """
    Clasifica el sueldo según las reglas proporcionadas:
    a. Sueldo bajo: por debajo de 300 dólares.
    b. Sueldo normal: entre 300 a 900.
    c. Sueldo mejor de lo normal: más de 900 dólares.
    """
    if cobro < 300:
        return 'Bajo'
    elif 300 <= cobro <= 900:
        return 'Normal'
    else: 
        return 'Mejor de lo normal'
```

```python
# Aplicar la función a la columna 'Cobro_USD'

df['Clasificacion_Calculada'] = df['Cobro_USD'].apply(clasificar_sueldo)
```

```python
# Mostrar el resultado
print("--- DataFrame con la Clasificación de Sueldo ---")
print(df[['Mes', 'Cobro_USD', 'Clasificacion_Calculada']])
```

```python
print("\n--- Resumen de Clasificaciones ---")
# Muestra cuántos meses caen en cada categoría
print(df['Clasificacion_Calculada'].value_counts())
print(
    'A) Sueldo bajo: por debajo de 300 dólares.', 
    'B) Sueldo normal:  entre 300 a 900.', 
    'C) Sueldo mejor de lo normal: más de 900 dólares.'
     )
print(f"El sueldo promedio mensual de Juan fue de: ${Sueldo_promedio:.2f}")
```
